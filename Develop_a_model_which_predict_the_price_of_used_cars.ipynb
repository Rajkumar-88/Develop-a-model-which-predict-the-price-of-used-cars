**Encoding the categorial variables**

```python
dataset.select_dtypes(include='object').columns
```

```python
# one hot encoding
dataset = pd.get_dummies(data=dataset, drop_first=True)
```

```python
dataset.head()
```

```python
dataset.shape
```

**Splitting the dataset**

```python
dataset.head()
```

```python
# matrix of features
x=dataset.drop(columns='Selling_Price')
```

```python
# target variables
y=dataset['Selling_Price']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x, y,test_size=0.2, random_state=0)
```

```python
x_test.shape
```

**Building the model**

```python
from sklearn.ensemble import RandomForestRegressor
regressor= RandomForestRegressor()
regressor.fit(x_train, y_train)
```

```python
y_pred=regressor.predict(x_test)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
```

**Hyperparameter Tuning**

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
parameters = {
    'n_estimators' :[100, 200, 300, 400, 500, 600, 700, 800, 900, 1000],
    'criterion' :['mse', 'mae'],
    'max_depth' :[10, 20, 30, 40, 50],
    'min_samples_split' :[2, 5, 10, 20, 50],
    'min_samples_leaf' :[1, 2, 5, 10],
    'max_features' :['auto','sqrt', 'log2']
}
```

```python
parameters
```

```python
random_cv = RandomizedSearchCV(estimator=regressor, param_distributions=parameters, n_iter=10,
                               scoring='neg_mean_absolute_error', cv=5, verbose=2, n_jobs=-1)
```
